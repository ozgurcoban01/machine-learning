```python
timesData = pd.read_csv("data/timesData.csv")
```

```python
timesData.head()
```

```python
dataframe = timesData[timesData.year == 2015]
```

```python
trace1 = go.Scatter3d(
    x=dataframe.world_rank,  #x eksenine dnya sralamas eklenir.
    y=dataframe.research,    #y eksenine research skoru eklenir.
    z=dataframe.citations,   #z eksenine citations skoru eklenir.
    mode='markers',          #marker -> nokta demektir.
    marker=dict(
        size=dataframe.teaching, #size eklenerek 4 boyutlu yaplr.
        color='rgb(120,0,0)',    #renk eklenir.
    )
)

data = [trace1]  #trace1 listenin ierisine koyulur.

layout = go.Layout(
    margin=dict(  #kenarlardan braklan boluktur.
        l=0,
        r=0,
        b=0,
        t=0       #hepsi 0 verildii iin kenarlardan boluk braklmaz.
    )
    
)
fig = go.Figure(data=data, layout=layout)  #data parametresine data deikeni, layout parametresine layout deikeni atanarak figr oluturulur.

iplot(fig) #figr plot edilir.
```
