```python
timesData = pd.read_csv("data/timesData.csv")
```

```python
timesData.head()
```

```python
df2011 = timesData[timesData.year == 2011].iloc[:6,:] 
df2011
```

```python
#create trace1
trace1 = go.Bar( #Bar tr seilir.
                x = df2011.university_name, #x eksenine niversite isimleri koyulur.
                y = df2011.citations,  #y eksenine alnt deerleri koyulur.
                name = "citations",    #trace'in ad verilir.
                marker = dict(color = 'rgba(240, 240, 240, 0.8)', #renk ve saydamlk verilir.
                             line=dict(color='rgb(0,0,0)',width=1.5)), #bar plotun evresindeki renk ve evresinin kalnl
                text = df2011.country) #deerin zerine gelindiinde hoverda lke ad yazr.

# create trace2 
trace2 = go.Bar(
                x = df2011.university_name,  #x eksenine niversite isimleri koyulur.
                y = df2011.teaching,    #y eksenine retme deerleri koyulur.
                name = "teaching",      #trace'in ad verilir.
                marker = dict(color = 'rgba(255, 255, 128, 0.5)',   #renk ve saydamlk verilir.
                              line=dict(color='rgb(0,0,0)',width=1.5)), #bar plotun evresindeki renk ve evresinin kalnl
                text = df2011.country)   #deerin zerine gelindiinde hoverda lke ad yazr.

data = [trace1, trace2] #trace1 ve trace2 data deikenine liste olarak atanr.

layout = go.Layout(barmode = "group") #barmode -> trace1 ve trace2'nin nasl koyulacann modudur.
                                      #group moduyla iki trace yan yana koyulur.

fig = go.Figure(data = data, layout = layout) #data parametresine data deikeni, layout parametresine layout deikeni atanarak figr oluturulur.

iplot(fig) #figr plot edilir.
```

```python
x = df2011.university_name

trace1 = { #burada bir dictionary yaratlr.
  'x': x, #x eksenine niversite ad koyulur.
  'y': df2011.citations, #y eksenine alnt deerleri koyulur.
  'name': 'citation', #trace'in ad verilir.
  'type': 'bar'  #plot tr belirlenir.
};
trace2 = {
  'x': x,  #x eksenine niversite ad koyulur.
  'y': df2011.teaching,  #y eksenine alnt deerleri koyulur.
  'name': 'teaching',   #trace'in ad verilir.
  'type': 'bar'  #plot tr belirlenir.
};

data = [trace1, trace2]; #trace1 ve trace2 data deikenine liste olarak atanr.

layout = {
  'xaxis': {'title': 'Top 3 universities'},
  'barmode': 'relative',  #trace1 ve trace2 alt alta koyulur.
  'title': 'citations and teaching of top 6 universities in 2011'
};

fig = go.Figure(data = data, layout = layout) #data parametresine data deikeni, layout parametresine layout deikeni atanarak figr oluturulur.

iplot(fig) #figr plot edilir.
```
