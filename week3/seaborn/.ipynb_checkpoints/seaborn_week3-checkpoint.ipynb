```python
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np

data = sns.load_dataset("flights") 
data = data.pivot("month", "year", "passengers")


sns.heatmap(data=data, annot=True, fmt="d",cmap=sns.cubehelix_palette(as_cmap=True))

plt.show()
```

```python
data = pd.DataFrame({
    'age_group': np.repeat(['18-30', '31-45', '46-60', '61+'], 100),
    'feedback_rating': np.random.randint(1, 6, 400)
})

g = sns.FacetGrid(data, col="age_group", margin_titles=True)

# Grafik tipi
g.map(sns.histplot, "feedback_rating")
```

```python

sns.set_theme(style="white")

df = sns.load_dataset("penguins")

g = sns.JointGrid(data=df, x="body_mass_g", y="bill_depth_mm", space=0)
g.plot_joint(sns.kdeplot,
             fill=True, clip=((2200, 6800), (10, 25)),
             thresh=0, levels=100, cmap="crest")
g.plot_marginals(sns.histplot, color="#03051A", alpha=1, bins=25)
```
